```python
import numpy as np

from torch import nn
import torch
from torchvision.utils import draw_keypoints, draw_bounding_boxes
```

```python
from torchvision import transforms
import torchvision.datasets

data = torchvision.datasets.Kitti(
    root="./data", train=True, download=True, transform=transforms.ToTensor()
)
```

```python
# show first image:
import matplotlib.pyplot as plt

# matplotlib enlarge image display size:
plt.figure(figsize=(10, 10))

idx = 3

img = draw_bounding_boxes(
    data[idx][0],
    torch.tensor([x["bbox"] for x in data[idx][1]]),
    width=5,
    labels=[x["type"] for x in data[idx][1]],
    font_size=24,
)

plt.imshow(img.permute(1, 2, 0));
```
